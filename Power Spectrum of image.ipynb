```python
import cv2
import numpy as np
import radialProfile
import pylab as py
path = "Dead-leaves-test-target.jpg"
# path = "Function1/gaussian-noise.png"
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
print(image.shape)

# Take the fourier transform of the image.
F1 = np.fft.fft2(image)

# Now shift the quadrants around so that low spatial frequencies are in
# the center of the 2D fourier transformed image.
F2 = np.fft.fftshift( F1 )

# Calculate a 2D power spectrum
psd2D = np.abs( F2 )**2

# Calculate the azimuthally averaged 1D power spectrum
psd1D = radialProfile.azimuthalAverage(psd2D)

# Now plot up both
py.figure(1)
py.clf()
py.imshow( image , cmap=py.cm.Greys)

py.figure(2)
py.clf()
py.imshow( np.log10( image ), cmap=py.cm.Greys)

py.figure(3)
py.clf()
py.imshow( np.log10( psd2D ))

py.figure(4)
py.clf()
py.semilogy( psd1D )
py.xlabel("Spatial Frequency")
py.ylabel("Power Spectrum")

py.show()
```

 the power spectrum of the dead leaves target follows a power law
